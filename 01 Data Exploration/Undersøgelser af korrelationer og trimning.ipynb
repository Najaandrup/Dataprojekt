```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("data_values.tsv", sep = '\t')
data["signal"] = data["signal"].apply(lambda x: np.array([int(i) for i in x.split(",")]))
```

### PolyALength vs AdapterLength

```python
plt.figure(figsize=(8, 5))

for i in range(10):
    polya_length = data["polya_end"][i] - data["polya_start"][i]
    adapter_length = data["polya_start"][i]
    plt.plot(polya_length, adapter_length, marker = 'o')

plt.xlabel("PolyALength")
plt.ylabel("AdapterLength")
plt.show()
```

### PolyALength vs BodyLength

```python
plt.figure(figsize=(8, 5))

for i in range(10):
    polya_length = data["polya_end"][i] - data["polya_start"][i]
    body_length = len((data["signal"][i])) - data["polya_end"][i]
    plt.plot(polya_length, body_length,marker = "o")

plt.xlabel("PolyALength")
plt.ylabel("BodyLength")
plt.show()
```

### PolyALength vs (Randomly) Trimmed AdapterLength

```python
plt.figure(figsize=(8, 5))

for i in range(10):
    polya_length = data["polya_end"][i] - data["polya_start"][i]
    adapter_length = data["polya_start"][i] - 350
    plt.plot(polya_length, adapter_length, marker = 'o')

plt.xlabel("PolyALength")
plt.ylabel("AdapterLength")
plt.show()
```
