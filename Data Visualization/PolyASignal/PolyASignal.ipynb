```python
import pandas as pd
import pod5 as p5
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
tsv_file = "egfp_a60_30_polyA_position.tsv"
pod_file = "egfp_a60_30_polya_reads.pod5"

polyA_df = pd.read_csv(tsv_file, sep='\t')

with p5.Reader(pod_file) as reader:
    pod_data = [(str(read.read_id), read.signal, read.pore) for read in reader.reads()]
pod_df = pd.DataFrame(pod_data, columns=["read_id", "signal", "pore"])

df = polyA_df.merge(pod_df, on="read_id")
df["signal"] = df.apply(lambda row: row["signal"][row["start"]:row["end"]], axis=1)

all_signals = np.concatenate(df["signal"].values)

plt.figure(figsize=(10, 6))
plt.hist(all_signals, bins=100, color="blue", alpha=0.7, edgecolor="black")

plt.xlabel("Signal Intensity")
plt.ylabel("Frequency")
plt.xlim(0, 2000)
plt.ylim(0, 3.5e8)
plt.title("Distribution of Signal Intensities - A60-30")

plt.show()
```
